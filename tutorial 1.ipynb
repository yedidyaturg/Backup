```python
import numpy as np
import cvxpy as cp
```

# (d) 
# Explicit representation


```python
x = cp.Variable(2)
objective = cp.Maximize(7*x[0] + 10*x[1])
constraints = [3*x[0] + 2*x[1] <= 36, 2*x[0] + 4*x[1] <= 40, x[0] >= 0, x[1] >= 0]
prob = cp.Problem(objective, constraints)
sol = prob.solve(solver = 'GUROBI')

print(f'Optimal solution is: x* = {x.value.T}')
print(f'Optimal value is: {sol}')
```

# Matrix representation


```python
c = np.array([[7], [10]])
A = np.block([[3,2], [2, 4], [-1,0], [0,-1]])
b = np.array([[36], [40], [0], [0]])
```

```python
x = cp.Variable((2,1))
objective = cp.Maximize(c.T@x)
constraints = [A @ x <= b]
prob = cp.Problem(objective, constraints)
sol = prob.solve(solver = 'GUROBI')

print(f'Optimal solution is: x* = {x.value.T}')
print(f'Optimal value is: {sol}')
```
